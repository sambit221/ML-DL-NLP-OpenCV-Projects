```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
df.shape #42000 rows and 785 columns
# 1st col is label and rest 784 are pixels of a greyscale image
```

```python
df.head(n=4)
```

```python
# converting dataframe to numpy array
data=df.values
np.random.shuffle(data)
print(type(data))
print(data.shape)
```

```python
# separating data and label
x = data[:, 1:]
y = data[:,0] #labels

print(x.shape,y.shape)
```

```python
# plotting images using the provided data
def drawImg(x, y, i):
    plt.imshow(x[i].reshape(28,28)) # reshaping 784 to 28*28
    plt.title("Label "+ str(y[i]))
    plt.show()
    
# drawImg(x,y,5)
for j in range(5):
    drawImg(x,y,j)
```

```python
split=int(0.80*x.shape[0])
print(split)
```

```python
x_train, y_train = x[: split,:],y[:split] # upto the split
x_test, y_test = x[split:,:],y[split:] # starting from the split to the last
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
# plotting a visualisation
plt.figure(figsize=(8,8))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x_train[i].reshape(28,28))
    plt.title(y_train[i])
    plt.axis("off")
```
